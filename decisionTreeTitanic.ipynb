```python
import pandas as pd
```

```python
df=pd.read_csv("F:/MachineLearning/titanic.csv")
df
```

```python
final=df.drop(["PassengerId", "Name", "SibSp", "Parch", "Ticket", "Cabin", "Embarked"],axis="columns")
final
```

```python
#dummies=pd.get_dummies(df[["Car Model"]])
#dummies
```

```python
final.isnull().sum()
```

```python
final=final.dropna()
final
```

```python
dummies=pd.get_dummies(final["Sex"])
dummies=dummies.drop(["male"],axis=1)
dummies
```

```python
merged=pd.concat([final,dummies],axis=1)
merged
```

```python
merged=merged.drop(["Sex"],axis=1)
merged
```

```python
X=merged.drop(["Survived"],axis=1)
X
```

```python
y=merged.drop(["Pclass", "Age", "Fare", "female"],axis=1)
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test= train_test_split(X,y)
```

```python
X_train
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(random_state=0)
```

```python
model.fit(X,y)
```

```python
model.predict([[2,47.0,10.5000,0]])
```

```python
model.score(X_test,y_test)
```
